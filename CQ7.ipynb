for: se usa para iterar sobre una secuencia (lista, tupla, etc.).

```
for i in range(3):
    print(i)
```
def: sirve para definir funciones en python.
```
def prg():
    print("como estás?")
prg()
```
return: devuelve un valor desde una función.
```
def r(a, b):
    return a - b
print(r(6, 9))
```
while: ejecuta un bloque de código mientras una condición sea verdadera.
```
c = 0
while c < 3:
    print(c)
    c += 1
```
True y False: son constantes booleanas que representas que algo es verdadero o falso.

```
x = True
y = False
print(x and not y)
```
None: es u valor ausente o nulo.
```
x = None
if x is None:
    print("Nulo")
```
And, or y not: son operadores logicos.

```
x, y = True, False
print(x and y)
print(x or y)  
print(not x)
```
if, elif, else: so estructuras condicionales.
```
x = 20
if x > 10:
    print("Mayor")
elif x == 20:
    print("Igual")
else:
    print("Menor")
```
break: rompe un bucle. No es tan recomendable usarlo.
```
for i in range(8):
    if i == 2:
        break
    print(i)
```
continue: salta a la siguiente iteracion del bucle
```
for i in range(5):
    if i == 2:
        continue
    print(i)
```
pass: instrucción nula, es decir no hace nada.
```
def funcion():
    pass
```
Try, except, finally, raise, assert: sirven para manejar los errores.
```
try:
    x = 80 / 0
except ZeroDivisionError:
    print("Error: división entre cero")
finally:
    print("Siempre se ejecuta")
```

lambda: es una función anónima, acorta el proceso.
```
cuadrado = lambda p: p**2
print(cuadrado(4))
```
Global y nonlocal: acceso a variables de ámbito superior.
```
v = 10
def f():
    global v
    v = 90
f()
print(v)
```
```
def externo():
    v = 5
    def interno():
        nonlocal v
        v = 10
    interno()
    print(v)  # 90
externo()
```
class: sirve para definir una clase

```
class Persona:
    def __init__(self, n):
        self.n = n

p = Persona("Pedro")
print(p.n)
```

in: comprueba que un elemento si está en la secuencia.
```
print(2 in [1, 2, 3, 4])
```
with: ayuda a abrir archivos.
```
with open("archivo.txt", "w") as f:
    f.write("Bienvenido")
```
yield: convierte una dfunción en un generador.
```
def c():
    for i in range(5):
        yield i
for n in c():
    print(n)
```
