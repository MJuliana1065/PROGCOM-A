```python
import tkinter as tk
from tkinter import messagebox
from datetime import datetime


def calcular_pam(sistolica, diastolica):
    pam = (sistolica + 2 * diastolica) / 3
    return pam


def calcular():
    try:
        sistolica = float(entry_sistolica.get())
        diastolica = float(entry_diastolica.get())

        pam = calcular_pam(sistolica, diastolica)
        resultado_label.config(text=f"La presión arterial media es: {pam:.2f} mmHg")

        if pam < 60:
            estado = "PAM baja: riesgo de hipoperfusión."
            color = "red"
        elif pam > 100:
            estado = "PAM elevada: posible hipertensión."
            color = "orange"
        else:
            estado = "PAM dentro del rango normal."
            color = "green"

        estado_label.config(text=estado, fg=color)

        guardar_historial(sistolica, diastolica, pam, estado)

    except ValueError:
        messagebox.showerror("Error", "Por favor, ingresa valores numéricos válidos.")


def guardar_historial(sistolica, diastolica, pam, estado):
    fecha = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    with open("historial_pam.txt", "a") as f:
        f.write(f"{fecha} | Sistólica: {sistolica} | Diastólica: {diastolica} | PAM: {pam:.2f} | {estado}\n")


def mostrar_historial():
    try:
        with open("historial_pam.txt", "r") as f:
            contenido = f.read()
        if contenido:
            messagebox.showinfo("Historial de Mediciones", contenido)
        else:
            messagebox.showinfo("Historial de Mediciones", "No hay mediciones registradas aún.")
    except FileNotFoundError:
        messagebox.showinfo("Historial de Mediciones", "No hay mediciones registradas aún.")

ventana = tk.Tk()
ventana.title("Calculadora de Presión Arterial Media (PAM)")
ventana.geometry("400x350")
ventana.resizable(False, False)

tk.Label(ventana, text="Presión Sistólica (mmHg):", font=("Arial", 12)).pack(pady=5)
entry_sistolica = tk.Entry(ventana, font=("Arial", 12))
entry_sistolica.pack()

tk.Label(ventana, text="Presión Diastólica (mmHg):", font=("Arial", 12)).pack(pady=5)
entry_diastolica = tk.Entry(ventana, font=("Arial", 12))
entry_diastolica.pack()

tk.Button(ventana, text="Calcular PAM", command=calcular, font=("Arial", 12), bg="#4CAF50", fg="white").pack(pady=10)
tk.Button(ventana, text="Ver historial", command=mostrar_historial, font=("Arial", 12), bg="#2196F3", fg="white").pack(pady=5)

resultado_label = tk.Label(ventana, text="", font=("Arial", 12))
resultado_label.pack(pady=5)

estado_label = tk.Label(ventana, text="", font=("Arial", 12, "bold"))
estado_label.pack(pady=5)

ventana.mainloop()
```
